```python
# Librerías
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
import os
```

```python
# Opciones de navegación
options =  webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')

driver_path = 'C:\\Users\\Santoyle\\Downloads\\chromedriver.exe'

driver = webdriver.Chrome(driver_path, chrome_options=options)
```

```python
df = pd.read_csv('razsocv2.csv')
df
```

```python
df2 = pd.DataFrame(columns = ['RS','CUIT'])
```

```python
from selenium.common.exceptions import TimeoutException
for row, datos in df.iterrows():
    
    empresas = datos['Empresas']
    try :
        driver.get('https://www.cuitonline.com/')
    
        WebDriverWait(driver, 2)\
        .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                      'input#searchBox')))\
        .send_keys(empresas)
    
        WebDriverWait(driver, 2)\
        .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                      'button.button search-button-home'.replace(' ', '.'))))\
        .click()
    
        WebDriverWait(driver, 3)\
        .until(EC.element_to_be_clickable((By.XPATH,
                                      '/html/body/div[1]/section/div/div[1]/div[2]/div[2]/div[2]/span[1]/span[2]')))
    
        cuit = driver.find_element_by_xpath('/html/body/div[1]/section/div/div[1]/div[2]/div[2]/div[2]/span[1]/span[2]')
        cuit = cuit.text
        print(cuit)
        #for i in cuit:
        df2.loc[len(df2)] = [empresas, cuit]
        
    
    except TimeoutException as e:
        df2.loc[len(df2)] = [empresas, 'No tiene CUIT!!']
        print("No tiene CUIT!!")
    
driver.quit()

df2.to_csv('CUITs.csv')
```
